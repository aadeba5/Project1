# Silos vs. Product Category Comparisons - Jillian

```python
#Sales Analysis Data Frame
product_category_count = len(sales_data['code_cat_prdt'].unique())
silo_count = len(sales_data['Silo'].unique())
item_count = len(sales_data['id_item'].unique())
total_qty = sales_data['qty_ship'].sum()
total_sales = sales_data['price_net'].sum()


sales_analysis = pd.DataFrame([{'Total Product Categories': product_category_count,
                               'Total Silos': silo_count,
                               'Number of Unique Items': item_count,
                               'Qty Shipped':total_qty,
                               'Total Sales': total_sales}])
sales_analysis
```

```python
silos = sales_data.groupby(['Silo_Description'])
silo_sales = silos.sum()
silo_sales
```

```python
sales_by_silo = silo_sales['price_net']
sales_by_silo

pd.DataFrame(sales_by_silo)
```

```python
product_categories = sales_data.groupby(['category'])
category_sales = product_categories.sum()
category_sales
```

```python
sales_by_category = category_sales['price_net']
sales_by_category

pd.DataFrame(sales_by_category)
```

```python
silo_chart = sales_by_silo.plot(kind='bar', title='Sales by Silo')
plt.show()
```
